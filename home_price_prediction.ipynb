```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
df=pd.read_csv("homeprice.csv")
df
```

```python
%matplotlib inline
plt.title("prediction of home_price")
plt.xlabel("area(sqr ft)",fontsize=20)
plt.ylabel("price(us$)",fontsize=20)
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
#To make predict by linear regression
#we use the linear function

reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
# the equation is 
# y=mx+b
#here y is price , m is slope and  b is intercept

#we are good to predict the price

#predict the area of 3300 sqt ft

reg.predict([[3300]])
```

```python
#coef_ is slope of the lineae regression
reg.coef_
```

```python
#intercept is the value of b
reg.intercept_
```

```python
135.78767123*3300+180616.43835616432
```
